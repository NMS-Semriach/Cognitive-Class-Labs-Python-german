<h2 id="set">Sets</h2>

<h3 id="content">Set Inhalte</h3>

Eine Menge (*ein Set*) ist eine eindeutige Sammlung von Objekten in Python, dh in der Menge gibt es keine doppelten Werte. Du kannst ein Set mit einer geschweiften Klammer <b>{}</b> definieren. Python entfernt dabei automatisch doppelte Elemente:

```python
# Erstelle ein Set

set1 = {"pop", "rock", "soul", "hard rock", "rock", "R&B", "rock", "disco"}
set1
```

Der Abbildungsvorgang ist in der Grafik dargestellt:

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/SetsUnique.png" width="1100" />

Du kannst ein Set auch aus einer Liste wie folgt anlegen:

```python
# Umwandlung einer Liste in ein Set

album_list = [ "Michael Jackson", "Thriller", 1982, "00:42:19", \
              "Pop, Rock, R&B", 46.0, 65, "30-Nov-82", None, 10.0]
album_set = set(album_list)             
album_set
```

Und jetzt erstellen wir ein Set von Genres:

```python
# Umwandlung einer Liste in ein Set

music_genres = set(["pop", "pop", "rock", "folk rock", "hard rock", "soul", \
                    "progressive rock", "soft rock", "R&B", "disco"])
music_genres
```

<h3 id="op">Set-Operationen</h3> 

Betrachten wir die Set-Operationen, da diese zum Ändern der Sets verwendet werden können. Wir gehen von einem Set <b>A</b> aus:

```python
# Beispielset

A = set(["Thriller", "Back in Black", "AC/DC"])
A
```

Wir können ein Element zu einem Set mit der Methode <code>add()</code> hinzufügen: 

```python
# Füge einem Set ein Element hinzu

A.add("NSYNC")
A
```

Wenn wir das gleiche Element zweimal hinzufügen, passiert nichts, da es keine Duplikate in einem Set geben kann:


```python
# Versuchsweise ein doppeltes Element zur Menge hinzufügen.

A.add("NSYNC")
A
```

Wir können einen Eintrag aus einem Set mit der Methode <code>remove</code> entfernen:

```python
# Entfernen eines Elements aus dem Set

A.remove("NSYNC")
A
```

Wir können mit dem Befehl <code>in</code> überprüfen, ob ein Element Teil des Sets ist:

```python
# Überprüfung, ob sich das Element im Set befindet.

"AC/DC" in A
```

<h3 id="logic">Sets und Logik-Operationen</h3>

Beachte, dass du mit Sets den Unterschied zwischen den Sets sowie Symmetrieunterschiede, Schnittmengen und Vereinigungen abfragen kannst:

Betrachten wir die folgenden beiden Sets:

```python
# Beispielsets

album_set1 = set(["Thriller", 'AC/DC', 'Back in Black'])
album_set2 = set([ "AC/DC", "Back in Black", "The Dark Side of the Moon"])
```

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/SetsSamples.png" width="650" />

```python
# Ausgabe beider Sets

album_set1, album_set2
```

Da beide Sets <b>AC/DC</b> und <b>Back in Black</b> enthalten, beschreiben wir diese gemeinsamen Elemente mit der Überlappung von zwei Kreisen.

<img src = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/SetsLogic.png" width = "650" />

Du kannst die Schnittmenge von zwei Sets mit Hilfe von <code>&</code> wie folgt finden:

```python
# Ermittlung der Schnittmenge

intersection = album_set1 & album_set2
intersection
```

Du kannst alle Elemente, die nur in <code>album_set1</code> enthalten sind, mit der Methode <code>difference</code> finden:

```python
# Feststellung der Differenz zwischen Set1 im Vergleich zu Set2

album_set1.difference(album_set2)  
```

Du musst nur Elemente in <code>album_set1</code> berücksichtigen; alle Elemente in <code>album_set2</code>, einschließlich der Schnittmenge, sind nicht enthalten.

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/SetsLeft.png" width="650" />

Die Elemente in <code>album_set2</code> die aber nicht in <code>album_set1</code> enthalten sind werden folgendermaßen ermittelt:

```python
album_set2.difference(album_set1)  
```

<img src = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/SetsRight.png" width="650" />

Du kannst auch die Schnittmenge von <code>album_list1</code> und <code>album_list2</code> auch mit der Methode <code>intersection</code> finden:

```python
# Mit der Methode intersection kann man die Schnittmenge von album_list1 und album_list2 ermitteln.

album_set1.intersection(album_set2)   
```

Dies entspricht der Überlappung der beiden Kreise:

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/SetsIntersect.png" width="650" />

Die Vereinigung entspricht allen Elementen in beiden Mengen, was durch das Einfärben beider Kreise dargestellt wird:

<img src = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/SetsUnion.png" width="650" />

Die Vereinigung erhält man mittels der Methode <code>union</code>:

```python
# Die Vereinigung von zwei Sets ermitteln

album_set1.union(album_set2)
```

Und du kannst folgendermaßen überprüfen, ob ein Set eine Obermenge bzw. Teilmenge eines anderen Sets ist:

```python
# Überprüfung, ob Obermenge

set(album_set1).issuperset(album_set2)   
```

```python
A = {1,2,3,4,5}
B = {2,3}
A.issuperset(B)
B.issubset(A)
```

```python
# Teilmenge überprüfen

set(album_set2).issubset(album_set1)     
```

Hier ist ein Beispiel, bei dem <code>issubset()</code> und <code>issuperset()</code> true zurückgeben:

```python
# Teilmenge überprüfen

set({"Back in Black", "AC/DC"}).issubset(album_set1) 
```

```python
# Obermenge überprüfen

album_set1.issuperset({"Back in Black", "AC/DC"})   
```
