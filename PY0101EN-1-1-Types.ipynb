<h3 id="comments">Kommentare schreiben in Python</h3>

<p>
    Es ist immer eine gute Idee, nicht nur Code zu schreiben, sondern deinem Code auch Kommentare hinzuzufügen. Es wird anderen helfen zu verstehen, was du versucht hast zu erreichen (der Grund, warum du einen bestimmten Ausschnitt des Codes geschrieben hast). Dies hilft nicht nur <strong>anderen Leuten</strong>, deinen Code zu verstehen, es kann auch als Erinnerung dienen <strong>für dich</strong>, wenn du Wochen oder Monate später darauf zurückkommst.</p>

<p>
    Um Kommentare in Python zu schreiben, benutze das Zahlensymbol <code>#</code>, bevor du deinen Kommentar schreibst. Wenn du deinen Code ausführst, ignoriert Python alles, was hinter dem <code>#</code> in einer bestimmten Zeile steht.
</p>

```python
# Übung im Schreiben von Kommentaren

print('Hallo, Python!') # Diese Zeile gibt eine Zeichenkette aus.
# print('Hi')
```

<p>
    Nach der Ausführung der obigen Zelle solltest du beachten, dass <code>Diese Zeile gibt eine Zeichenkette aus</code> nicht in der Ausgabe erschienen ist, da es sich um einen Kommentar handelte (und somit von Python ignoriert wurde).
</p>
<p>
    Die zweite Zeile wurde auch nicht ausgeführt, weil <code>print('Hi')</code> ebenfalls mit dem Nummernzeichen (<code>#</code>) versehen war! Da es sich hierbei nicht um einen erklärenden Kommentar des Programmierers, sondern um eine tatsächliche Codezeile handelt, sagt man, dass der Programmierer diese zweite Zeile des Codes <em>auskommentiert hat</em>.
</p>

<h3 id="errors">Fehler (Errors) in Python</h3>

<p>Jeder macht Fehler. Für viele Arten von Fehlern wird Python dir sagen, dass du einen Fehler gemacht hast, indem es dir eine Fehlermeldung ausgibt. Es ist wichtig, die Fehlermeldungen sorgfältig zu lesen, um wirklich zu verstehen, wo du einen Fehler gemacht hast und wie du ihn korrigieren kannst.</p>
<p>Zum Beispiel, wenn du <code>print</code> als <code>frint</code> schreibst, wird Python eine Fehlermeldung anzeigen. Probieren es aus:</p>

```python
# Print falsch geschrieben erzeugt eine Fehlermeldung

frint("Hallo, Python!")
```

<p>Die Fehlermeldung sagt dir in diesem Fall: 
<ol>
    <li>wo (hier: in Zeile 2) der Fehler aufgetreten ist (nützlicher bei großen Notebookzellen oder Skripten), und</li>
    <li>welcher Art von Fehler es war (hier: NameError)</li>
</ol>
<p>Hier hat Python hat versucht, die Funktion <code>frint</code> auszuführen, konnte aber nicht feststellen, was <code>frint</code> ist, da es sich um keine eingebaute Funktion handelt und sie auch nicht vorher von uns definiert wurde.</p>

<p>
    Du wirst feststellen, dass, wenn wir einen anderen Fehler machen, indem wir vergessen, die Zeichenkette zu schließen, wir einen anderen Fehler erhalten (z.B. einen <code>SyntaxError</code>). Probiere es unten aus:
</p>

```python
# Teste die eingebaute Fehlermeldung.

print("Hallo, Python!)
```

<h3 id="python_error">Erkennt Python deinen Fehler, bevor es deinen Code ausführt?</h3>

Kompilierte Sprachen untersuchen dein gesamtes Programm zur Kompilierzeit und sind in der Lage, dich vor der Ausführung über eine ganze Reihe von Fehlern zu informieren. Im Gegensatz dazu interpretiert Python dein Skript Zeile für Zeile, während es ausgeführt wird. Python hört auf, das gesamte Programm auszuführen, wenn es auf einen Fehler stößt (es sei denn, der Fehler wird vom Programmierer erwartet und behandelt, einem fortgeschritteneren Thema, das wir später in diesem Kurs behandeln werden).

<h3 id="convert">Konvertierung von einem Objekttyp in einen anderen</h3>

<p>Du kannst den Typ des Objekts in Python ändern; dies wird als Typecasting bezeichnet. Du kannst beispielsweise einen <i>Integer</i>-Wert in einen <i>Float</i>-Wert (z.B. 2 in 2.0) umwandeln (konvertieren).</p>
<p>Probieren wir es aus:</p>

```python
# Überprüfung, ob es sich um eine Ganzzahl (Integer) handelt.

type(2)
```

<h4>Umwandlung (Casting) von ganzen Zahlen in Gleitkommazahlen</h4>
<p>Wir möchten den Integerwert 2 in einen Floatwert umwandeln:</p>

```python
# Konvertiere 2 zu float

float(2)
```

```python
# Konvertiere 2 (integer) zu float und überprüfe diesen Wert

type(float(2))
```

<p>Wenn wir eine ganze Zahl in einen Float umwandeln, ändern wir nicht wirklich den Wert (d.h. die Genauigkeit) der Zahl. Wenn wir jedoch einen Float in eine Ganzzahl umwandeln, könnten wir möglicherweise einige Informationen verlieren. Wenn wir zum Beispiel den Float 1.1 nach Integer umwandeln, erhalten wir 1 und verlieren die Dezimalinformationen (d.h. 0.1):</p>

```python
# Das Casting von 1.1 in eine ganze Zahl führt zum Verlust von Informationen.

int(1.1)
```

<h3 id="var">Variablen</h3>

<p>Wie bei den meisten Programmiersprachen können wir Werte in <i>Variablen</i> speichern, damit wir sie später verwenden können. Zum Beispiel:</p>

```python
# Wert in Variable speichern

x = 43 + 60 + 16 + 41
```

<p>Um den Wert von <code>x</code> in unserem Notebook zu sehen, können wir ihn einfach auf die letzte Zeile einer Zelle setzen:</p>

<p>Wenn wir einen Wert in einer bestehenden Variablen speichern, überschreibt der neue Wert den vorherigen Wert:</p>

```python
# Überschreiben der Variablen mit einem neuen Wert

x = x / 60
x
```

<p>Es ist empfehlenswert, sinnvolle Variablennamen zu verwenden, damit du und andere den Code lesen und leichter verstehen können:</p>

```python
# Bezeichne die Variablen sinnvoll

gesamt_min = 43 + 42 + 57 # Gesamtlänge der Musikalben in Minuten
gesamt_min
```

```python
# Bezeichne die Variablen sinnvoll

gesamt_stunden = gesamt_min / 60 # Gesamtlänge der Musikalben in Stunden
gesamt_stunden
```

<p>In den obigen Zellen haben wir die Länge von drei Alben in Minuten hinzugefügt und in <code>gesamt_min</code> gespeichert. Wir haben den Wert dann durch 60 geteilt, um die Gesamtlänge <code>gesamt_stunden</code> in Stunden zu berechnen. Du kannst das auch alles auf einmal in einem einzigen Ausdruck erledigen, solange du die Länge des Albums in Klammern setzt, bevor du teilst, wie unten gezeigt.</p>

```python
# Komplizierterer Ausdruck

gesamt_stunden = (43 + 42 + 57) / 60  # Gesamtstunden in einem einzigen Ausdruck
gesamt_stunden
```

<p>Wenn du lieber die Gesamtstunden als Integer haben möchtest, kannst du natürlich die Gleitkomma-Division durch eine Integer-Division ersetzen (z.B., <code>//</code>).</p>
