<h2 id="Dic">Dictionaries</h2>

<h3 id="content">Was sind Dictionaries?</h3>

Ein Dictionary besteht aus Schlüsseln (*Keys*) und Werten (*Values*). Es ist hilfreich, ein Dictionary mit einer Liste zu vergleichen. Statt der numerischen Indizes, wie bei z.B. einer Liste, haben Dictionaries Keys. Diese Keys sind die Schlüssel, mit denen auf Werte innerhalb eines Dictionary zugegriffen wird.

<img src="http://march.nmssemriach.at/pythonkurs/bilder/dictslist.png" width="650" />

Ein Dictionarybeispiel <code>Dict</code>:

```python
# Erstellen eines Dictionary

Dict = {"key1": 1, "key2": "2", "key3": [3, 3, 3], "key4": (4, 4, 4), ('key5'): 5, (0, 1): 6}
Dict
```

Die Keys (*Schlüssel*) können Strings sein:

```python
# Zugriff auf den Wert über den Key

Dict["key1"]
```

Keys kann auch ein beliebiges unveränderliches Objekt wie z.B. ein Tupel sein:  

```python
# Zugriff auf den Wert über den Key

Dict[(0, 1)]
```

Jeder Key wird durch einen Doppelpunkt "<code>:</code>" von seinem Wert getrennt.  Beistriche trennen die Elemente, und das ganze Dictionary ist in geschweiften Klammern eingeschlossen. Ein leeres Dictionary ohne Einträge wird mit nur zwei geschweiften Klammern geschrieben, so wie dieses "<code>{}</code>".

```python
# Erstellen eines Beispieldictionarys

release_year_dict = {"Thriller": "1982", "Back in Black": "1980", \
                    "The Dark Side of the Moon": "1973", "The Bodyguard": "1992", \
                    "Bat Out of Hell": "1977", "Their Greatest Hits (1971-1975)": "1976", \
                    "Saturday Night Fever": "1977", "Rumours": "1977"}
release_year_dict
```

Zusammenfassend lässt sich sagen, dass ein Dictionary wie eine Liste eine Reihe von Elementen enthält. Jedes Element wird durch einen Key und den entsprechenden Wert repräsentiert. Dictionaries werden mit zwei geschweiften Klammern erstellt, die Keys und Werte enthalten, die durch einen Doppelpunkt getrennt sind. Für jeden Key kann es nur einen einzigen Wert geben, aber mehrere Keys können den gleichen Wert enthalten. Keys können nur Strings, Zahlen oder Tupel sein, aber Werte können einen beliebigen Datentyp haben.

Es ist hilfreich, das Dictionary als Tabelle darzustellen, wie im folgenden Bild. Die erste Spalte repräsentiert die Keys, die zweite Spalte die Werte.

<h3 id="key">Keys</h3>

Du kannst die Werte über die Bezeichnungen abrufen:

```python
# Wert über Key abrufen

release_year_dict['Thriller'] 
```

Das entspricht:

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/DictsKeyOne.png" width="500" />

Ebenso für <b>The Bodyguard</b>.

```python
# Wert über Key abrufen

release_year_dict['The Bodyguard'] 
```

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/DictsKeyTwo.png" width="500" />

Du kannst auch alle Keys des Dictionary *release_year_dict* mit der Methode <code>keys()</code> abrufen:

```python
# Alle Keys des Dictionary abrufen

release_year_dict.keys()
```

Du kannst die Werte mit der Methode <code>values()</code> abrufen:

```python
# Alle Werte des Dictionary abrufen
release_year_dict.values() 
```

Wir können einen Eintrag hinzufügen:

```python
# Wert mit Key in Dictionary anhängen

release_year_dict['Graduation'] = '2007'
release_year_dict
```

Wir können einen Eintrag löschen:   

```python
# Einträge über den Key löschen

del(release_year_dict['Thriller'])
del(release_year_dict['Graduation'])
release_year_dict
```

 Wir können überprüfen, ob sich ein Element im Dictionary befindet: 

```python
# Prüfen ob der Schlüssel im Dictionary ist.

'The Bodyguard' in release_year_dict
```
