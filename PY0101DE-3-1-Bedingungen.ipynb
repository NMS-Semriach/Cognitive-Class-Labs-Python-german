<h3 id="comp">Vergleichsoperatoren</h3>

Vergleichsoperationen vergleichen einen Wert oder Operanden und erzeugen, basierend auf einer Bedingung, ein Boolean. Beim Vergleich zweier Werte kannst du folgende Operatoren verwenden:

<ul>
    <li>gleich: <b>==</b></li>
    <li>ungleich: <b>!=</b></li>
    <li>größer als: <b>></b></li>
    <li>kleiner als: <b>&lt;</b></li>
    <li>größer gleich: <b>>=</b></li>
    <li>kleiner gleich: <b>&lt;=</b></li>
</ul>

Wir weisen <code>a</code> einen Wert von 5 zu. Benutze den Gleichheitsoperator, der mit zwei Gleichheitszeichen <b>==</b> angegeben wird, um festzustellen, ob zwei Werte gleich sind. Der folgende Fall vergleicht die Variable <code>a</code> mit 6.

```python
# Bedingung Gleich

a = 5
a == 6
```

Das Ergebnis ist <b>False</b>, da 5 nicht gleich 6 ist.

Betrachten Sie den folgenden Gleichheitsvergleichsoperator <code>i > 5</code>. Wenn der Wert des linken Operanden, in diesem Fall die Variable <b>i</b>, größer ist als der Wert des rechten Operanden, in diesem Fall 5, dann ist die Aussage <b>True</b>. Andernfalls ist die Aussage <b>False</b>.  Wäre <b>i</b> gleich 6, wäre die Aussage <b>True</b>, weil 6 größer als 5 ist.

```python
# Größer als Zeichen

i = 6
i > 5
```

Weise i den Wert 2 zu, <code>i = 2</code> . Die folgende Aussage ist dann falsch, da 2 nicht größer als 5 ist:

```python
# Größer als Zeichen

i = 2
i > 5
```

Wir wollen jetzt einige Werte für <code>i</code> in der Grafik anzeigen, wobei für Werte größer als 5 der Hintergrund grün und für die anderen rot sein soll. Der grüne Bereich stellt dar, wo die obige Bedingung **True** ist, der rote, wo die Aussage **False** ist. Wenn der Wert von <code>i</code> 2 ist, erhalten wir **False**, da die 2 in den roten Bereich fällt. Wenn der Wert für <code>i</code> 6 ist, erhalten wir entsprechend **True**, da die Bedingung in den grünen Bereich fällt. 

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%203/Images/CondsGreater.gif" width="650" />

Die Ungleichheitsprüfung verwendet ein Ausrufezeichen vor dem Gleichheitszeichen, und wenn zwei Operanden ungleich sind, dann wird die Bedingung **True**.  Die folgende Bedingung ergibt beispielsweise **True**, solange der Wert von <code>i</code> nicht gleich 6 ist:

```python
# Ungleichheitszeichen

i = 2
i != 6
```

Wenn <code>i</code> gleich 6 ist, gibt die Ungleichheitsaussage <b>False</b> zurück. 

```python
# Inequality Sign

i = 6
i != 6
```

Betrachten wir die untenstehende Zahlenreihe. Wenn die Bedingung **True** ist, werden die entsprechenden Zahlen grün markiert und für den Fall, dass die Bedingung **False** ist, wird die entsprechende Zahl rot markiert.  Wenn wir <code>i</code> gleich 2 einstellen, ist das Ergebnis **True**, da 2 im grünen Bereich liegt. Wenn wir <code>i</code> gleich 6 setzen, erhalten wir als Ergebnis **False**, da die Bedingung in den roten Bereich fällt.

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%203/Images/CondsIneq.gif" width="650" />

Wir können die gleichen Methoden auf Strings anwenden. Wenn du z.B. einen Gleichheitsoperator für zwei verschiedene Strings verwendest erhalten wir **False** da die Strings nicht gleich sind.

```python
# Benutze das Gleichheitszeichen, um die Strings zu vergleichen.

"ACDC" == "Michael Jackson"
```

Wenn wir den Ungleichheitsoperator verwenden, wird die Ausgabe **True** sein, da die Strings nicht gleich sind.

```python
# Verwenden des Ungleichheitszeichens zum Vergleichen der Strings

"ACDC" != "Michael Jackson"
```

Die Ungleichheitsoperation wird auch verwendet, um die Buchstaben/Wörter/Symbole entsprechend dem ASCII-Wert von Buchstaben zu vergleichen. Der in der folgenden Tabelle dargestellte Dezimalwert stellt die Ordnungszahl des Zeichens dar:

So ist beispielsweise der ASCII-Code für <b>!</b> 21, während der ASCII-Code für <b>+</b> 43 ist. Daher ist <b>+</b> größer als <b>!</b>, da 43 größer als 21 ist.

Der Wert für <b>A</b> ist 101, und der Wert für <b>B</b> ist 102, weshalb auch hier gilt:

```python
# Zeichen vergleichen

'B' > 'A'
```

Wenn es mehrere Buchstaben gibt, hat der erste Buchstabe Vorrang bei der Reihenfolge:

```python
# Zeichen vergleichen

'BA' > 'AB'
```

<b>Hinweis</b>: Großbuchstaben haben einen anderen ASCII-Code als Kleinbuchstaben, d.h. der Vergleich zwischen den Buchstaben in Python ist abhängig von der Groß-/Kleinschreibung.

<h3 id="branch">Verzweigungen</h3>

Eine Verzweigung ermöglicht es uns, unterschiedliche Anweisungen für verschiedene Eingangsgrößen auszuführen. Es ist hilfreich, sich eine **if-Anweisung** als einen abgeschlossenen Raum vorzustellen. Wenn die Anweisung **True** ist, können wir den Raum betreten und dein Programm wird einige vordefinierte Aufgaben ausführen, aber wenn die Anweisung **False** ist, wird das Programm diese Aufgaben ignorieren.


Nehmen wir zum Beispiel das blaue Rechteck, das ein ACDC-Konzert darstellt. Wenn eine Person älter als 18 Jahre ist, kann sie am ACDC-Konzert teilnehmen. Wenn sie 18 oder jünger als 18 Jahre ist, kann sie nicht an dem Konzert teilnehmen.

Benutze die zuvor erlernten Bedingungsanweisungen, da die Bedingungen in der **if-Anweisung** überprüft werden müssen. Die Syntax ist so einfach wie <code> if <i>Bedingungsanweisung</i> :</code>, die das Wort <code>if</code>, eine beliebige Bedingungsanweisung und einen Doppelpunkt am Ende enthält. Die zu erledigende Aufgabe beginnt unter dieser Bedingung in einer neuen Zeile mit einem Einzug. Die Codezeilen nach dem Doppelpunkt und mit einem Einzug werden nur ausgeführt, wenn die **if-Anweisung** **True** ist. Die Aufgaben enden, wenn die Codezeile keinen Einzug mehr hat.

Im folgenden Fall wird die Aufgabe <code>print("Du kannst eintreten")</code> nur ausgeführt, wenn die Variable <code>age</code> größer als 18 ist und somit zu einer wahren Aussage führt, da diese Codezeile einen Einzug hat. Die Ausführung von <code>print("weiter geht's")</code> wird jedoch nicht durch die if-Anweisung beeinflusst.

```python
# Beispiel für eine If-Anweisung

alter = 19
#alter = 18

# Ausdruck, der wahr oder falsch sein kann.
if alter > 18:
    
    #innerhalb eines Einzugs steht die auszuführende Aufgabe, für den Fall dass die Bedingung wahr ist.
    print("Du kannst eintreten" )

#Die Anweisungen nach der if-Anweisung werden unabhängig davon ausgeführt, ob die Bedingung wahr oder falsch ist. 
print("weiter geht's")
```

<i>Versucht, die Variable <code>alter</code> auszukommentieren.

Es ist hilfreich, das folgende Diagramm zu verwenden, um den Prozess zu veranschaulichen. Auf der linken Seite sehen wir, was passiert, wenn die Bedingung <b>True</b> ist.  Die Person geht in das ACDC-Konzert, welches dem Code in dem gerade ausgeführten Einzug entspricht, und fährt dann fort. Auf der rechten Seite sehen wir, was passiert, wenn die Bedingung <b>False</b> ist; der Person wird kein Zugang gewährt, und die Person geht weiter. In diesem Fall wird das Codesegment im Einzug nicht ausgeführt, aber der Rest der Anweisungen sehr wohl.

Die Anweisung <code>else</code> führt einen Codeblock aus, wenn keine der Bedingungen vor dieser <code>else</code>-Anweisung  **True** ist. Laß uns noch einmal die ACDC-Konzertanalogie verwenden. Wenn der Benutzer 17 Jahre alt ist, kann er nicht zum ACDC-Konzert gehen, aber er kann zu Meat Loaf gehen.
Die Syntax der <code>else</code>-Anweisung ist ähnlich wie die Syntax der <code>if</code>-Anweisung, also <code>else:</code>. Beachte, dass es keine Bedingungsanweisung für <code>else</code> gibt.
Versuche, die Werte von <code>age</code> zu ändern, um zu sehen, was passiert:  

```python
# Else Anweisung Beispiel

alter = 18
# alter = 19

if alter > 18:
    print("Du kannst eintreten" )
else:
    print("schau dir Meat Loaf an" )
    
print("weiter geht's")
```

Der Ablauf wird im Folgenden demonstriert, wobei alle Möglichkeiten auf jeder Seite des Bildes dargestellt sind. Auf der linken Seite ist der Fall dargestellt, dass jemand 17 Jahre alt ist, wir setzen die Variable <code>alter</code> also auf 17, und das entspricht einer Person, die das Meat Loaf Konzert besucht. Der rechte Teil zeigt, was passiert, wenn die Person über 18 Jahre alt ist. In diesem Fall ist sie 19 Jahre alt, und sie darf zum ACDC Konzert.

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%203/Images/CondsElse.gif" width="650" />

Die <code>elif</code> Anweisung, kurz für else if, erlaubt es uns, zusätzliche Bedingungen zu überprüfen, wenn die Bedingungsanweisungen vor ihr <b>False</b> sind. Wenn die Bedingung für die <code>elif</code>-Anweisung <b>True</b> lautet, wird die alternative Angabe ausgeführt. Denke an das Konzertbeispiel, wo die Person, wenn sie 18 Jahre alt ist, zum Pink Floyd-Konzert geht, anstatt am ACDC- oder Meat-Loaf-Konzert teilzunehmen. Die Person im Alter von 18 Jahren betritt das Areal, und da sie nicht älter als 18 Jahre ist, kann sie ACDC nicht sehen, aber da sie 18 Jahre alt ist, besucht sie Pink Floyd. Nachdem sie Pink Floyd gesehen hat, geht es weiter. Die Syntax der <code>elif</code> Anweisung ist nicht neu, da wir lediglich für die <code>if</code>-Abfrage das <code>if</code> durch <code>elif</code> ersetzen müssen.

```python
# Beispiel für eine Elif-Anweisung

alter = 18

if alter > 18:
    print("Du kannst eintreten" )
elif alter == 18:
    print("schau dir Pink Floyd an")
else:
    print("schau dir Meat Loaf an" )
    
print("weiter geht's")
```

The three combinations are shown in the figure below.  The left-most region shows what happens when the individual is less than 18 years of age. The central component shows when the individual is exactly 18. The rightmost shows when the individual is over 18.

<h3 id="logic">Logik-Operatoren</h3>


Manchmal muss man mehr als eine Bedingung auf einmal überprüfen. Beispielsweise kannst du überprüfen, ob eine Bedingung und eine andere Bedingung **True** ist. Logische Operatoren ermöglichen es dir, Bedingungen zu kombinieren oder zu ändern.
<ul>
    <li><code>and</code></li></li>
    <li><code>or</code></li></li>
    <li><code>not</code></li></li>
</ul>

Diese Operatoren werden für zwei Variablen anhand der folgenden Wahrheitstabellen zusammengefasst:  

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%203/Images/CondsTable.png" width="650" />

Die Anweisung <code>and</code> ist nur **True**, wenn beide Bedingungen erfüllt sind. Die Anweisung <code>or</code> ist wahr, wenn eine Bedingung **True** ist. Die Anweisung <code>not</code> gibt den jeweils entgegengesetzten Wahrheitswert aus.

Schauen wir uns an, wie wir feststellen können, ob ein Album nach 1979 (1979 ist nicht enthalten) und vor 1990 (1990 ist nicht enthalten) veröffentlicht wurde. Die Zeiträume zwischen 1980 und 1989 erfüllen diese Bedingung. Dies wird in der folgenden Grafik veranschaulicht. Das Grün der Zeilen <strong>a</strong> und <strong>b</strong> repräsentiert Zeiträume, in denen die Aussage **True** ist. Das Grün auf der Linie <strong>c</strong> stellt dar, wo beide Bedingungen **True** sind, das entspricht dem Überlappungsbereich der oberen grünen Bereiche. 



<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%203/Images/CondsEgOne.png" width="650" />

Der Codeblock zur Ausführung dieser Überprüfung lautet:

```python
# Beispiel für eine Bedingungsanweisung

album_year = 1980

if(album_year > 1979) and (album_year < 1990):
    print ("Das Erscheinungsjahr liegt zwischen 1980 und 1989")
    
print("")
print("mach etwas ...")
```

Um festzustellen, ob ein Album vor 1980 (? - 1979) oder nach 1989 (1990 - ?) veröffentlicht wurde, kann eine **or** Anweisung verwendet werden. Zeiträume vor 1980 (? - 1979) oder nach 1989 (1990 - ?) erfüllen diese Bedingung. Dies wird in der folgenden Abbildung dargestellt, die Farbe Grün in <strong>a</strong> und <strong>b</strong> repräsentiert Zeiträume, in denen die Aussage wahr ist. Die Farbe Grün in **c** stellt dar, wo mindestens eine der Bedingungen wahr ist.  


<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%203/Images/CondsEgTwo.png" width="650" />

Der Codeblock zur Durchführung dieser Überprüfung lautet:

```python
# Beispiel für eine Bedingungsanweisung

album_year = 1990

if(album_year < 1980) or (album_year > 1989):
    print ("Das Album ist nicht in den 1980'ern erschienen")
else:
    print("Das Album erschien in den 1980'ern")
```

Die Anweisung <code>not</code> prüft, ob die Anweisung falsch ist:

```python
# Beispiel für eine Bedingungsanweisung

album_year = 1983

if not (album_year == '1984'):
    print ("Das Album ist nicht 1984 erschienen")
```
