<h2 id="tuple">Tupel</h2>

In Python gibt es verschiedene Datentypen: string, integer und float. Diese Datentypen können alle wie folgt in einem Tupel enthalten sein:

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/TuplesType.png" width="750" align="center" />

Nun, laß uns dein erstes Tupel mit einem String, Integer und Float erstellen.

```python
# Erstelle dein erstes Tupel

tuple1 = ("disco",10,1.2 )
tuple1
```

Der Typ der Variablen ist ein **Tupel**. 

```python
# Gibt den Typ des von dir erstellten Tupels aus.

type(tuple1)
```

<h3 id="index">Indexierung</h3>

Auf jedes Element eines Tupels kann über einen Index zugegriffen werden. Die folgende Tabelle stellt die Beziehung zwischen dem Index und den Elementen im Tupel dar. Jedes Element kann durch den Namen des Tupels erhalten werden, gefolgt von einer eckigen Klammer mit der Indexnummer:

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/TuplesIndex.gif" width="750" align="center">

Wir können jeden Wert im Tupel ausgeben:

```python
# Ausgabe der Variablen für jeden Index

print(tuple1[0])
print(tuple1[1])
print(tuple1[2])
```

Wir können den **Typ** jedes Wertes im Tupel ausgeben:

```python
# Ausgabe des Typs für jeden Index

print(type(tuple1[0]))
print(type(tuple1[1]))
print(type(tuple1[2]))
```

Wir können auch eine negative Indexierung verwenden, ausgehend vom letzten Element mit dem Index -1. Wir verwenden die gleiche Tabelle oben mit entsprechenden negativen Werten:

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/TuplesNeg.png" width="750" align="center">

Wir können das letzte Element also wie folgt erhalten (diesmal werden wir die Print-Anweisung nicht zur Anzeige der Werte verwenden):

```python
# Verwendung des negativen Index, um den Wert des letzten Elements zu erhalten.

tuple1[-1]
```

Wir können die nächsten beiden Elemente wie folgt anzeigen:

```python
# Verwendung des negativen Index, um den Wert des vorletzten Elements zu erhalten.

tuple1[-2]
```

```python
# Verwendung des negativen Index, um den Wert des vorvorletzten Elements zu erhalten.

tuple1[-3]
```

<h3 id="concate">Verkettung von Tupeln</h3>

Wir können Tupel verketten oder kombinieren, indem wir das **+** Zeichen verwenden:

```python
# Zusammenfügen von 2 Tupeln

tuple2 = tuple1 + ("hard rock", 10)
tuple2
```

Wir können Tupel aufteilen und erhalten mehrere Werte, wie die folgende Abbildung zeigt:

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/TuplesSlice.gif" width="750" align="center">

<h3 id="slice">Abschnitte</h3>

Wir können Tupel aufteilen und neue Tupel mit den entsprechenden Elementen erhalten: 

```python
# Aufteilung vom Index 0 bis Index 2

tuple2[0:3]
```

Wir können die letzten beiden Elemente des Tupels auslesen:

```python
# Abschnitt vom Index 3 bis Index 4

tuple2[3:5]
```

Wir können die Länge eines Tupels mit dem Befehl <code>len</code> ("length") ermitteln: 

```python
# Ermitteln der Länge des Tupels

len(tuple2)
```

Diese Abbildung zeigt die Anzahl der Elemente:

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/TuplesElement.png" width="750" align="center">

<h3 id="sort">Sortieren</h3>

Betrachte das folgende Tupel:

```python
# Beispiel eines Tupels

Bewertungen = (0, 9, 6, 5, 10, 8, 9, 6, 2)
```

Wir können die Werte in einem Tupel sortieren und in einem neuen Tupel speichern: 

```python
# Sortiere den Tupel

BewertungenSortiert = sorted(Bewertungen)
BewertungenSortiert
```

<h3 id="nest">Verschachtelte Tupel</h3>

Ein Tupel kann sowohl ein anderes Tupel als auch andere komplexere Datentypen enthalten. Dieser Prozess wird als "Verschachtelung" bezeichnet. Betrachte das folgende Tupel mit mehreren Elementen: 

```python
# Erstellen eines verschachtelten ("nested") Tupels

NestedT =(1, 2, ("pop", "rock") ,(3,4),("disco",(1,2)))
```

Jedes Element im Tupel einschließlich anderer Tupel kann über einen Index ermittelt werden, wie in der Abbildung dargestellt:

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/TuplesNestOne.png" width="750" align="center">

```python
# Gib das Element zu jedem Index aus

print("Element 0 des Tupels: ", NestedT[0])
print("Element 1 des Tupels: ", NestedT[1])
print("Element 2 des Tupels: ", NestedT[2])
print("Element 3 des Tupels: ", NestedT[3])
print("Element 4 des Tupels: ", NestedT[4])
```

Wir können den zweiten Index verwenden, um auf andere Tupel zuzugreifen, wie in der Abbildung gezeigt:

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/TuplesNestTwo.png" width="750" align="center">

 Wir können auf die verschachtelten Tupel zugreifen:

```python
# Ausgabe von Elementen für jeden Index, einschließlich geschachtelter Indizes

print("Element 2, 0 des Tupels: ",   NestedT[2][0])
print("Element 2, 1 des Tupels: ",   NestedT[2][1])
print("Element 3, 0 des Tupels: ",   NestedT[3][0])
print("Element 3, 1 des Tupels: ",   NestedT[3][1])
print("Element 4, 0 des Tupels: ",   NestedT[4][0])
print("Element 4, 1 des Tupels: ",   NestedT[4][1])
```

Wir können auf Strings in den zweiten verschachtelten Tupeln über einen dritten Index zugreifen:

```python
# Ausgabe des ersten Elements im zweiten verschachtelten Tupel

NestedT[2][1][0]
```

```python
# Ausgabe des zweiten Elements im zweiten verschachtelten Tupel

NestedT[2][1][1]
```

Wir können einen Baum verwenden, um den Prozess zu visualisieren. Jeder neue Index entspricht einer tieferen Ebene im Baum:

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/TuplesNestThree.gif" width="750" align="center">

Ebenso können wir auf Elemente zugreifen, die tiefer im Baum verschachtelt sind, mit einem vierten Index:

```python
# Ausgabe des ersten Elements im zweiten verschachtelten Tupel

NestedT[4][1][0]
```

```python
# Ausgabe des zweiten Elements im zweiten verschachtelten Tupel

NestedT[4][1][1]
```

Die folgende Abbildung zeigt die Beziehung des Baumes zu dem Element <code>NestedT[4][1][1][1]</code>:
